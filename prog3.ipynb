```python
import matplotlib.pyplot as plt
import numpy as np

# Increase figure size
plt.figure(figsize=(12, 12))

# Graph 1
x = np.array([2, 4])
y = np.array([3, 6])

plt.subplot(3, 3, 1)
plt.title('Title 1')
plt.xlabel("X label", color="blue")
plt.ylabel("Y label", color="blue")
plt.plot(x, y, 'o', color="blue")

# Graph 2
y1 = np.array([3, 8, 1, 9])
y2 = np.array([6, 2, 8, 3])

plt.subplot(3, 3, 2)
plt.plot(y1, color="orange", marker='*')
plt.plot(y2, linewidth=1.5, color="blue", linestyle="dashed")

# Graph 3
ypoints = np.array([3, 8, 1, 10])

plt.subplot(3, 3, 3)
plt.plot(ypoints, marker='o', ms=10, mec="black", mfc="yellow")

# Graph 4
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([210, 220, 230, 240, 250, 260, 270, 280, 290, 300])

plt.subplot(3, 3, 4)
plt.title("Title 2")
plt.plot(x, y)
plt.grid()

# Graph 5
x = np.random.randint(50, size=(50))
y = np.random.randint(50, size=(50))
colors = np.random.randint(50, size=(50))
sizes = 1 * np.random.randint(50, size=(50))

plt.subplot(3, 3, 5)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

# Graph 6
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.subplot(3, 3, 6)
plt.bar(x, y, color="red", width=0.1)

# Graph 7 - Pie Chart
plt.subplot(3, 3, 7)
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
# Plot pie chart
plt.pie(y, labels=mylabels, explode=myexplode, startangle=90, shadow=True, autopct='%1.1f%%')
# Place legend outside the pie chart
plt.legend(title="Four Fruits:", bbox_to_anchor=(1.05, 1), loc='upper left')

# Overall title
plt.suptitle("GRAPH", fontsize=20)

# Adjust subplot spacing
plt.subplots_adjust(left=0.1, right=0.8, top=0.9, bottom=0.1, wspace=0.3, hspace=0.3)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and set the size
plt.figure(figsize=(16, 4))

# Plot 1
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])
font1 = {'family': 'serif', 'color': 'blue', 'size': 20}
plt.subplot(1, 4, 1)
plt.plot(xpoints, ypoints, 'o')
plt.xlabel("x label", fontdict=font1)
plt.ylabel("y label", fontdict=font1)
plt.title("Title")

# Plot 2
x_m = [1, 2, 3, 4, 5]
y_m = [5, 3, 5, 3, 5]

x_w = [1, 2, 3, 4, 5]
y_w = [3, 5, 3, 5, 3]

plt.subplot(1, 4, 2)
plt.plot(x_m, y_m, marker='*', linestyle='-', color='orange', label='M shape')
plt.plot(x_w, y_w, marker='*', linestyle=':', color='blue', label='W Shape')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('M and W Shapes')
plt.legend()

# Plot 3
x_n = [0, 0.5, 0.5, 1]
y_n = [0, 1, 0, 1]

plt.subplot(1, 4, 3)
plt.plot(x_n[:-1], y_n[:-1], marker='o', markersize=10, markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='black', linestyle='-', color='black', linewidth=2)
plt.plot([x_n[-2], x_n[-1]], [y_n[-2], y_n[-1]], marker='o', markersize=10, markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='black', linestyle='-', color='blue', linewidth=2)

plt.xlim(-0.1, 1.1)
plt.ylim(-0.1, 1.1)
plt.gca().set_aspect('equal', adjustable='box')
plt.axis('off')

plt.title('N', fontsize=15)

# Plot 4
x = np.linspace(0, 10, 100)
y = x

plt.subplot(1, 4, 4)
plt.plot(x, y, label='y-x', color='blue')
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title 2', loc='left')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12, 12))

# Graph 1
x = np.array([1, 3])
y = np.array([4, 7])

plt.subplot(3, 3, 1)
plt.title('Graph 1')
plt.xlabel("X axis", color="green")
plt.ylabel("Y axis", color="green")
plt.plot(x, y, 's', color="green")

# Graph 2
y1 = np.array([2, 7, 3, 10])
y2 = np.array([5, 1, 9, 4])

plt.subplot(3, 3, 2)
plt.plot(y1, color="purple", marker='x')
plt.plot(y2, linewidth=2, color="red", linestyle="dotted")

# Graph 3
ypoints = np.array([2, 9, 4, 12])

plt.subplot(3, 3, 3)
plt.plot(ypoints, marker='s', ms=12, mec="blue", mfc="green")

# Graph 4
x = np.array([70, 75, 80, 85, 90, 95, 100, 105, 110, 115])
y = np.array([200, 210, 220, 230, 240, 250, 260, 270, 280, 290])

plt.subplot(3, 3, 4)
plt.title("Graph 2")
plt.plot(x, y)
plt.grid()

# Graph 5
x = np.random.randint(30, size=(50))
y = np.random.randint(30, size=(50))
colors = np.random.randint(30, size=(50))
sizes = 5 * np.random.randint(30, size=(50))

plt.subplot(3, 3, 5)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')

# Graph 6
x = np.array(["W", "X", "Y", "Z"])
y = np.array([4, 9, 2, 11])

plt.subplot(3, 3, 6)
plt.bar(x, y, color="cyan", width=0.2)

# Graph 7 - Pie Chart
plt.subplot(3, 3, 7)
y = np.array([30, 20, 30, 20])
mylabels = ["Pears", "Plums", "Peaches", "Nectarines"]
myexplode = [0.1, 0, 0, 0]
plt.pie(y, labels=mylabels, explode=myexplode, startangle=120, shadow=True, autopct='%1.1f%%')
plt.legend(title="Fruit Types:", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.suptitle("All Graphs", fontsize=20)

plt.subplots_adjust(left=0.1, right=0.8, top=0.9, bottom=0.1, wspace=0.3, hspace=0.3)

plt.show()
```
