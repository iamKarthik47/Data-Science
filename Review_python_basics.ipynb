```python
# Numeric Data Types
int_var = 10
float_var = 10.5
complex_var = 3 + 4j

print(f"int_var: {int_var}, type: {type(int_var)}")
print(f"float_var: {float_var}, type: {type(float_var)}")
print(f"complex_var: {complex_var}, type: {type(complex_var)}")

# Mathematical Operations
int_var += 5
float_var *= 2
complex_var -= 1 + 1j

print(f"Updated int_var: {int_var}")
print(f"Updated float_var: {float_var}")
print(f"Updated complex_var: {complex_var}")

# Boolean Variables
bool_var1 = True
bool_var2 = False

print(f"bool_var1: {bool_var1}, type: {type(bool_var1)}")
print(f"bool_var2: {bool_var2}, type: {type(bool_var2)}")

# Boolean Operations
print(f"bool_var1 AND bool_var2: {bool_var1 and bool_var2}")
print(f"bool_var1 OR bool_var2: {bool_var1 or bool_var2}")
print(f"NOT bool_var1: {not bool_var1}")

# Strings
string_var1 = "Hello"
string_var2 = "World"

print(f"string_var1: {string_var1}, length: {len(string_var1)}")
print(f"string_var2: {string_var2}, length: {len(string_var2)}")

# String Concatenation and Formatting
concatenated_string = string_var1 + " " + string_var2
formatted_string = f"{string_var1}, {string_var2}!"

print(f"Concatenated string: {concatenated_string}")
print(f"Formatted string: {formatted_string}")

# String Methods
capitalized_string = concatenated_string.capitalize()
uppercase_string = concatenated_string.upper()
justified_string = concatenated_string.ljust(20)
centered_string = concatenated_string.center(20)
replaced_string = concatenated_string.replace("World", "Python")
stripped_string = "  Hello Python  ".strip()

print(f"Capitalized string: {capitalized_string}")
print(f"Uppercase string: {uppercase_string}")
print(f"Justified string: '{justified_string}'")
print(f"Centered string: '{centered_string}'")
print(f"Replaced string: {replaced_string}")
print(f"Stripped string: '{stripped_string}'")

# Containers: List, Tuple, Set, and Dictionary
list_var = [1, 2, 3, 4, 5]
tuple_var = (1, 2, 3, 4, 5)
set_var = {1, 2, 3, 4, 5}
dict_var = {"one": 1, "two": 2, "three": 3}

print(f"list_var: {list_var}, type: {type(list_var)}")
print(f"tuple_var: {tuple_var}, type: {type(tuple_var)}")
print(f"set_var: {set_var}, type: {type(set_var)}")
print(f"dict_var: {dict_var}, type: {type(dict_var)}")

# Functions
def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

def sub(a,b):
    return a-b
def div(a,b):
    return a/b

print(f"add(10, 20): {add(10, 20)}")
print(f"multiply(10, 20): {multiply(10, 20)}")

# Classes
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        return f"Hello, my name is {self.name} and I am {self.age} years old."

person = Person("Alice", 30)
print(person.greet())
```
